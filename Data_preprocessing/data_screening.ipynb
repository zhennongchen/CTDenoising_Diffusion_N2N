### screening the fixed static data we have (100 cases used in motion correction projection as static reference in the simulation)

```python
patient_sheet = pd.read_excel(os.path.join('/mnt/camca_NAS/Portable_CT_data/','Patient_list', 'NEW_CT_concise_collected_fixed_static_edited.xlsx'),dtype={'Patient_ID': str, 'Patient_subID': str})
patient_sheet['use'] = patient_sheet['use'].fillna(0)
patient_sheet = patient_sheet[(patient_sheet['use'] != 0) & (patient_sheet['use'] != 'no')]
print('patient sheet len: ', len(patient_sheet))
```

```python
img_shape_list = []
pixel_dim_list = []
for i in range(0,patient_sheet.shape[0]):
    row = patient_sheet.iloc[i]
    patient_id = row['Patient_ID']
    patient_subid = row['Patient_subID']

    patient_folder = os.path.join(main_path, 'nii_imgs_202404', 'static', patient_id, patient_subid, 'fixed')

    if os.path.isfile(os.path.join(patient_folder, 'img_1mm.nii.gz'))!= 1:
        print(patient_id, patient_subid)
    
    use = row['use']
    img_file = os.path.join(patient_folder, use +'.nii.gz')
    img = nb.load(img_file)

    # find the pixel dim and shape
    pixel_dim = img.header['pixdim'][1:4]
    img_shape = img.header['dim'][1:4]
    img_shape_list.append(img_shape)
    pixel_dim_list.append(pixel_dim)

    print('patient_id: ', patient_id, 'patient_subid: ', patient_subid, 'img_shape: ', img_shape, 'pixel_dim: ', pixel_dim)

# add two columns into patient sheet
patient_sheet['img_shape'] = img_shape_list
patient_sheet['pixel_dim'] = pixel_dim_list

# save
patient_sheet.to_excel(os.path.join('/mnt/camca_NAS/denoising/','Patient_lists', 'fixedCT_static.xlsx'), index = False)
```
